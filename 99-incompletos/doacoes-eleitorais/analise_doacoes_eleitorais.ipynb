# 2. Ler as bases de dados

```python
df_votos = pd.read_csv('votacao_candidato_SP.csv', encoding='latin-1', sep=';')

df_votos = df_votos.groupby(['Nome candidato']).agg({'Votos nominais':sum})
```

```python
df_receita = pd.read_csv('receitas_candidatos_SP.csv', encoding='latin-1', sep=';')
df_receita = df_receita[df_receita['Situação de candidatura']=='Apto']
df_receita = df_receita[['Nome candidato', 'Nome doador', 'Origem da receita', 'Fonte de receita', 'Partido', 'Situação de totalização', 'Valor de receita']]
```

```python

df_receita['Valor de receita'] = df_receita['Valor de receita'].str.replace(',', '.').astype(float)

df_candidato = df_receita.groupby('Nome candidato')['Valor de receita'].sum()

df_doadores = df_receita[df_receita['Origem da receita']=='Recursos de pessoas físicas'].groupby(['Nome doador']).agg({'Valor de receita':'sum'}).reset_index()
```
